```python
import numpy as np
import pandas as pd
from sklearn import datasets
data = pd.DataFrame(data=datasets.load_iris().data, columns=datasets.load_iris().feature_names)
# Add a column for the target variable (species)
data['target'] = datasets.load_iris().target
# Map the target values to their respective target names
data['target_names'] = datasets.load_iris().target_names[data.target]
data = data.drop("target", axis = 1)
data.shape
data.head()
y = data.iloc[:,4].values
print(y)
y = np.where(y=='setosa',-1,1)
print(y)
x = data.iloc[:,[1,2]].values
print(x)
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=99)
from sklearn.linear_model import Perceptron
p=Perceptron(max_iter=40,eta0=0.1,random_state=99)
p.fit(x_train,y_train)
y_predict=p.predict(x_test)
print(y_predict)
```

```python
from sklearn.metrics import accuracy_score 
print("Accuracy : %.2f"%accuracy_score(y_test,y_predict))
```
